```python
import numpy as np
data = np.loadtxt("backup", dtype=np.int)
type(data)
```

```python
k = len(data)
n = 1024;

result = []
for i in range(n):
    if i in data:
        continue
        
    result.append(i)

lst = list(data)
lst.reverse()
result += lst
#print(result)
np.savetxt(str(n) + "_" + str(k) + ".txt", result, fmt="%d", delimiter=' ', newline=' ');
#result[500]
```
