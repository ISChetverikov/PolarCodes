```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

sc_df = pd.read_csv("..//results//SC_clean.results", skipinitialspace=True, sep=',', comment='#')
scfano_df = pd.read_csv("..//results//SCFano_clean.results", skipinitialspace=True, sep=',',comment='#')

snr = sc_df['SNR']
sc = sc_df['FER']
snr2 = scfano_df['SNR']
fano = scfano_df['FER']
plt.figure(figsize=(10,8))
plt.plot(snr, sc, '-o', snr, fano, '-D');
plt.title("SC vs SCFano")
plt.legend(["SC", "SCFano"])
plt.xlabel("SNR")
plt.grid(True)
plt.ylabel("FER")
plt.yscale('log')
plt.savefig("ScVsScFano.png");
```
