```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

scl1_df = pd.read_csv("..//results//SCL1.results", skipinitialspace=True, sep=',', comment='#')
scl2_df = pd.read_csv("..//results//SCL2.results", skipinitialspace=True, sep=',', comment='#')
scl4_df = pd.read_csv("..//results//SCL4.results", skipinitialspace=True, sep=',', comment='#')
scl8_df = pd.read_csv("..//results//SCL8.results", skipinitialspace=True, sep=',', comment='#')
scl16_df = pd.read_csv("..//results//SCL16.results", skipinitialspace=True, sep=',', comment='#')
scf1_df = pd.read_csv("..//results//SCFLIPPROG_488_level1_.results", skipinitialspace=True, sep=',', comment='#')
scf2_df = pd.read_csv("..//results//SCFLIPPROG_488_level2_copy.results", skipinitialspace=True, sep=',', comment='#')
scf3_df = pd.read_csv("..//results//SCFLIPPROG_488_level3_.results", skipinitialspace=True, sep=',', comment='#')


snr = scl1_df['SNR']
scl1 = scl1_df['FER']
scl2 = scl2_df['FER']
scl4 = scl4_df['FER']
scl8 = scl8_df['FER']
snr_l8 = scl8_df['SNR']

scl16 = scl16_df['FER']
snr_l16 = scl16_df['SNR']

scf1 = scf1_df['FER']
scf2 = scf2_df['FER']
scf3 = scf3_df['FER']


plt.figure(figsize=(10,8))
plt.plot(snr, scl1, snr, scl2, snr, scl4, snr_l8, scl8, snr_l16, scl16, snr, scf1,snr, scf2, snr, scf3);
plt.title("SC vs SCFlip")
plt.legend(["SC", "SCL2", "SCL4", "SCL8", "SCL16", "SCFlipProgLevel1","SCFlipProgLevel2","SCFlipProgLevel3"])
plt.xlabel("SNR")
plt.grid(True)
plt.ylabel("FER")
plt.yscale('log')
plt.savefig("ScVsScFanoNew.png");
```
