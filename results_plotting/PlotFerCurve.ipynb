```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

sc_df = pd.read_csv("..//results//SC_clean.results", skipinitialspace=True, sep=',', comment='#')
scfano_df = pd.read_csv("..//results//SCFano_clean.results", skipinitialspace=True, sep=',',comment='#')
scfano3_df = pd.read_csv("..//results//SCFano_clean3.results", skipinitialspace=True, sep=',',comment='#')

#display(scfano3_df)

snr = sc_df['SNR']
sc = sc_df['FER']
snr2 = scfano_df['SNR']
fano = scfano_df['FER']
fano3 = scfano3_df['FER']
plt.figure(figsize=(10,8))
plt.plot(snr, sc, '-o', snr, fano, snr, fano3, '-D');
plt.title("SC vs SCFano")
plt.legend(["SC", "SCFano", "SCFanoNew"])
plt.xlabel("SNR")
plt.grid(True)
plt.ylabel("FER")
plt.yscale('log')
plt.savefig("ScVsScFanoNew.png");
```
