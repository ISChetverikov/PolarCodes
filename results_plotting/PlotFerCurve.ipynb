```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

scl1_df = pd.read_csv("..//results//SCL1.results", skipinitialspace=True, sep=',', comment='#')
scl2_df = pd.read_csv("..//results//SCL2.results", skipinitialspace=True, sep=',', comment='#')
scl4_df = pd.read_csv("..//results//SCL4.results", skipinitialspace=True, sep=',', comment='#')
scf1_df = pd.read_csv("..//results//SCProgFlip1024_488_level1.results", skipinitialspace=True, sep=',', comment='#')
scf2_df = pd.read_csv("..//results//SCProgFlip1024_488_level2.results", skipinitialspace=True, sep=',', comment='#')


snr = scl1_df['SNR']
scl1 = scl1_df['FER']
scl2 = scl2_df['FER']
scl4 = scl4_df['FER']
scf1 = scf1_df['FER']
scf2 = scf2_df['FER']


plt.figure(figsize=(10,8))
plt.plot(snr, scl1, snr, scl2, snr, scl4, snr, scf1,snr, scf2);
plt.title("SC vs SCFlip")
plt.legend(["SC", "SCL2", "SCL4", "SCFlipProgLevel1","SCFlipProgLevel2"])
plt.xlabel("SNR")
plt.grid(True)
plt.ylabel("FER")
plt.yscale('log')
plt.savefig("ScVsScFanoNew.png");
```
